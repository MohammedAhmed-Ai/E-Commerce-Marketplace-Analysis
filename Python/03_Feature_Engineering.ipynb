# E-Commerce Marketplace Customer Analytics

## Feature Engineering

This notebook creates additional features that enhance business analysis, SQL queries, Power BI dashboards, and future machine learning models.

### Objectives

- Create date-related features
- Calculate shipping metrics
- Create profit-related metrics
- Prepare analytical features

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
```

## 2. Load Clean Dataset

```python
dataset_path = "../Dataset/Processed/cleaned_sales.csv"

df = pd.read_csv(dataset_path)

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

print("Dataset loaded successfully.")
```

## 3. Date Features

```python
df["Order Year"] = df["Order Date"].dt.year
```

```python
df["Order Month"] = df["Order Date"].dt.month
```

```python
df["Order Month Name"] = df["Order Date"].dt.month_name()
```

```python
df["Order Quarter"] = df["Order Date"].dt.quarter
```

```python
df["Order Day"] = df["Order Date"].dt.day
```

```python
df["Order Weekday"] = df["Order Date"].dt.day_name()
```

## 4. Shipping Features

```python
df["Shipping Days"] = (df["Ship Date"] -df["Order Date"]).dt.days
```

```python
df["Shipping Days"].describe()
```

## 5. Profit Features

```python
df["Profit Margin"] = (df["Profit"] / df["Sales"]) * 100
```

```python
df["Profit Margin"] = (df["Profit Margin"].replace([np.inf, -np.inf], np.nan))
```

```python
df["Profit Margin"].describe()
```

## 6. Discount Features

```python
df["Has Discount"] = np.where(df["Discount"] > 0,"Yes","No")
```

```python
median_profit = df["Profit"].median()

df["High Profit"] = np.where(df["Profit"] >= median_profit,"High","Low")
```

```python
df["Order Size"] = np.where(df["Sales"] >= df["Sales"].median(),"Large","Small")
```

## 8. Save Engineered Dataset

```python
output_path = "../Dataset/Processed/featured_sales.csv"

df.to_csv(output_path, index=False)

print("Feature engineering completed successfully.")
```
