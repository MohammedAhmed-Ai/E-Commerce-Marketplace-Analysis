# E-Commerce Marketplace Customer Analytics

## Data Cleaning

This notebook focuses on cleaning and preparing the raw dataset for SQL analysis, Power BI visualization, and machine learning.

### Objectives
- Correct data types
- Handle missing values
- Validate data consistency
- Prepare a clean dataset

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
```

## 2. Load Raw Dataset

```python
dataset_path = "../Dataset/Raw/Sales.csv"

df = pd.read_csv(dataset_path, encoding="latin1")

print("Dataset loaded successfully.")
```

## 3. Convert Data Types

```python
df.dtypes
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"],dayfirst=True)

df["Ship Date"] = pd.to_datetime(df["Ship Date"],dayfirst=True)
```

```python
df[["Order Date", "Ship Date"]].dtypes
```

## 4. Handle Missing Values

```python
df.isnull().sum()
```

```python
df["Postal Code"] = (df["Postal Code"].fillna("Unknown").astype(str))
```

```python
df["Postal Code"].isnull().sum()
```

```python
df["Postal Code"].dtype
```

## 5. Data Validation

```python
df[["Sales", "Quantity", "Discount", "Profit"]].describe().T
```

```python
negative_sales = df[df["Sales"] < 0]

print(f"Negative Sales Records: {len(negative_sales)}")
negative_sales.head()
```

```python
zero_sales = df[df["Sales"] == 0]

print(f"Zero Sales Records: {len(zero_sales)}")
zero_sales.head()
```

```python
invalid_discount = df[
    (df["Discount"] < 0) |
    (df["Discount"] > 1)
]

print(f"Invalid Discount Records: {len(invalid_discount)}")
```

```python
invalid_quantity = df[df["Quantity"] <= 0]

print(f"Invalid Quantity Records: {len(invalid_quantity)}")
```

### Validate Date Columns

```python
invalid_dates = df[df["Ship Date"] < df["Order Date"]]

print(f"Invalid Date Records: {len(invalid_dates)}")
invalid_dates.head()
```

```python
df["Order ID"].duplicated().sum()
```

```python
df[df["Order ID"] == df["Order ID"].iloc[0]]
```

```python
df["Customer ID"].isnull().sum()
```

```python
df["Product ID"].isnull().sum()
```

## Validation Findings

### Validation Summary

- Date columns were successfully converted to datetime format.
- Postal Code was converted to string and missing values were replaced with "Unknown".
- No duplicate transaction records were found.
- No negative sales values were detected.
- Discount values fall within the expected range (0–1).
- Quantity values are valid.
- Shipping dates occur after or on the corresponding order dates.
- Customer IDs and Product IDs contain no missing values.

The dataset passed all validation checks and is ready for feature engineering and SQL analysis.

## 6. Save Clean Dataset

```python
output_path = "../Dataset/Processed/cleaned_sales.csv"

df.to_csv(output_path, index=False)

print("Clean dataset saved successfully!")
```
