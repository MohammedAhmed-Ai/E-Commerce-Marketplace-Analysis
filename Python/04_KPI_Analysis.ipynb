# E-Commerce Marketplace Customer Analytics

## KPI Analysis

This notebook calculates the key business metrics required for business reporting, SQL implementation, Power BI dashboards, and customer analytics.

### Objectives

- Sales KPIs
- Customer KPIs
- Product KPIs
- Regional KPIs

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
```

## 2. Load Featured Dataset

```python
dataset_path = "../Dataset/Processed/featured_sales.csv"

df = pd.read_csv(dataset_path)

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

print("Dataset loaded successfully.")
```

## 3. Sales KPIs

```python
total_sales = df["Sales"].sum()

print(f"Total Sales : ${total_sales:,.2f}")
```

```python
total_profit = df["Profit"].sum()

print(f"Total Profit : ${total_profit:,.2f}")
```

```python
total_orders = df["Order ID"].nunique()

print(f"Total Orders : {total_orders}")
```

```python
df["Sales"].mean()
```

```python
aov = total_sales / total_orders

print(f"Average Order Value : ${aov:,.2f}")
```

## 4. Customer KPIs

```python
total_customers = df["Customer ID"].nunique()

print(f"Total Customers : {total_customers}")
```

```python
repeat_customers = (df.groupby("Customer ID")["Order ID"].nunique())

repeat_customers = (repeat_customers > 1).sum()

print(f"Repeat Customers : {repeat_customers}")
```

```python
orders_per_customer = (total_orders /total_customers)

print(f"Average Orders per Customer : {orders_per_customer:.2f}")
```

## 5. Product KPIs

```python
top_products = (df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
)

top_products
```

```python
top_categories = (df.groupby("Category")["Sales"].sum().sort_values(ascending=False))

top_categories
```

```python
top_subcategories = (df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10))

top_subcategories
```

## 6. Regional KPIs

```python
sales_region = (df.groupby("Region")["Sales"].sum().sort_values(ascending=False))

sales_region
```

```python
profit_region = (df.groupby("Region")["Profit"].sum().sort_values(ascending=False))

profit_region
```

```python
top_states = (df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10))

top_states
```

## 7. Shipping KPIs

```python
average_shipping = (df["Shipping Days"].mean())

print(f"Average Shipping Days : {average_shipping:.2f}")
```

```python
shipping_mode = (df.groupby("Ship Mode")["Shipping Days"].mean().sort_values())

shipping_mode
```
