# Initial Findings & Data Assessment

### Key Findings

- The dataset contains 9,994 sales transaction records and 21 columns.
- No duplicate records were found.
- Only the `Postal Code` column contains missing values (11 records, approximately 0.11%).
- `Order Date` and `Ship Date` are currently stored as text and will be converted to datetime during the data cleaning phase.
- Profit contains both positive and negative values, which is expected in retail sales due to discounts and operational costs.
- The dataset includes 794 unique customers, 1,862 products, and 5,010 unique orders.
- Each order may contain multiple products, which explains why the number of transaction records is greater than the number of unique orders.
- The dataset is suitable for customer analytics, sales analysis, churn prediction, and business intelligence reporting.
